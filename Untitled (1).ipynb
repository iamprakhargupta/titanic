## Was gender a factor which effected the survival of a person? 



To answer this question we are going to find the Pearson's R using the dataframe.corr() function.Then we will plot the bar graph to compare between survival of males and females. To plot the bar graph we will use the titanic_df dataframe apply the vectorized operation to fill in a database bar which will be used for our computation
We will also find the survivability using the formula 

**No of people survived of a sex x 100 / total survivors**


```python
corr1=titanic_df[['sexbinary','Survived']].corr()
print(corr1.loc['sexbinary','Survived'])
```

The pearson's R is -0.54 the minus sign id due to the fact we assumed male to be 1 and female to be zero.
This result shouws that if your are a male ,its tough luck for you to survive.**Correspondingly if you are a woman you chance of survival are much higher**

**Lets calculate survivability for both male and woman for the titanic data using the formula we discussed earlier**

```python
bar=pd.DataFrame([[0,0],[0,0]], index=['survived','dead'],columns=['male','female'])
sex=pd.Series(titanic_df['Sex'])
sur=pd.Series(titanic_df['Survived'])
survivors=sex[sur==1]
dead=sex[sur==0]
for s in survivors:
    if s=='male':
        bar.loc['survived','male']+=1
    if s=='female':
        bar.loc['survived','female']+=1
for d in dead:
    if d=='male':
        bar.loc['dead','male']+=1
    if d=='female':
        bar.loc['dead','female']+=1
print(bar)
total=bar.loc['survived'].sum()
print('Male survivability {}%'.format( (bar.loc['survived','male']*100)/total))
print('Female survivability {}%'.format( (bar.loc['survived','female']*100)/total))
```

there is more than double chance of a female to survive titanic compared to a male

To make the bar graph we are going to use the dataframe.plot() function with kind argument to bar

```python
ax=bar.plot(kind='bar')#bar chart
ax.set_ylabel('Number of People')
ax.set_title('Comparison between survival of males and females')
```

## How did Socio-Economic Status have an impact on survival Status?

## Was embarked station a factor of survival?

To answer this question we will create a function droup with takens in a value and compare with embarkments 
C-Cherbourg,Q-Queenstown,S-Southampton
the function group perform a vectorized operation 
We have removed the 2 missing entries from our data by passing C Q S value to the group function. The group() return a series 

We have created a dataframe bar2 and merge the series using the .concat() function

Further we will calculate the survive percent for each embarkments by using this formula

**Survivors of a particular embarked station x 100 / Total number of people boarded from that embarked station**


```python
#function that group the people from embarked station together and return the  
def group(value): 
    Embarked=pd.Series(titanic_df['Embarked'])
    temp=titanic_df[Embarked==value] #vectorised operation
    return temp.groupby(['Survived','Sex']).count()['PassengerId']
   
```

```python
c=group('C')
q=group('Q')
s=group('S')
listof =[c,q,s]
c1=pd.DataFrame(c)


bar2=pd.concat([c1,q,s],keys=['Cherbourg','Queenstown','Southampton'],axis=1) #concating dataframe with series
bar2=bar2.rename(columns={'PassengerId':''})
bar2=bar2.rename(index={0:'Dead'})
bar2=bar2.rename(index={1:'Survived'})

ax=bar2.plot(kind='bar',colormap='Set2',stacked=True)
ax.set_ylabel('No of people')
ax.set_xlabel('Survial Status and Sex')
ax.set_title('Embarkment and Status of the whole dataset')
plt.legend(['Cherbourg','Queenstown','Southampton'])
```
